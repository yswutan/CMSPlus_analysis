```python
# =======================
# Common dependencies
# =======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, auc
from math import pi

plt.rcParams["font.family"] = "Arial"
plt.rcParams["svg.fonttype"] = "none"

```

```python
# Figure 5E â€” Immunity Response Barplot
data = {
    'Response': ['iPR', 'iSD', 'iUPD', 'iCPD'],
    'CMS1': [3, 7, 1, 0],
    'CMS2': [3, 7, 4, 4],
    'CMS3': [5, 9, 1, 5],
    'CMS4-TME+': [10, 7, 1, 4],
    'CMS4-TME-': [7, 13, 1, 5],
}
df = pd.DataFrame(data)

for c in ['CMS1', 'CMS2', 'CMS3', 'CMS4-TME+', 'CMS4-TME-']:
    df[f'Freq_{c}'] = df[c] / df[c].sum()

fig, ax = plt.subplots(figsize=(4, 4))
bar_width = 0.15
index = np.arange(len(df['Response']))
colors = ['#E69F00', '#377EB8', '#CC7AA7', '#3D5588', '#8491B5']

for i, col in enumerate(['Freq_CMS1','Freq_CMS2','Freq_CMS3','Freq_CMS4-TME+','Freq_CMS4-TME-']):
    ax.bar(index + i*bar_width, df[col], bar_width, label=col.replace('Freq_', ''), color=colors[i], alpha=0.6)

ax.set_xlabel('Response category', fontsize=14)
ax.set_ylabel('Response (rel. freq.)', fontsize=14)
ax.set_xticks(index + bar_width*2)
ax.set_xticklabels(df['Response'])
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig('img_plot/immunity_freq.png', dpi=300)
plt.savefig('img_plot/immunity_freq.svg', bbox_inches='tight')
plt.savefig('img_plot/immunity_freq.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
