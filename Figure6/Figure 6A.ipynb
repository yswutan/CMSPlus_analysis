```python
# =======================
# Common dependencies
# =======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, auc
from math import pi
plt.rcParams["font.family"] = "Arial"
plt.rcParams["svg.fonttype"] = "none"
```

```python
# Figure 5A â€” ROC Curve
def plot_roc_curve(data_path, model_name, color, alpha=1.0):
    """Read CSV with FPR/TPR and plot ROC curve."""
    data = pd.read_csv(data_path)
    all_fpr, all_tpr = data['fpr'].values, data['tpr'].values
    all_auc = auc(all_fpr, all_tpr)
    plt.plot(all_fpr, all_tpr, color=color, lw=2, alpha=alpha,
             label=f"{model_name} AUC = {all_auc:.3f}")
```

```python
plt.figure(figsize=(4, 3.6))
plot_roc_curve('table/Resnet18_overall_ROC_data.csv', 'ResNet-18', '#1F77B4', 0.5)
plot_roc_curve('table/Radiomics_overall_ROC_data.csv', 'Radiomics', '#ffc477', 0.8)
plot_roc_curve('table/Ours_overall_ROC_data.csv', 'CT4CMSPlus', '#f08877', 1.0)

plt.plot([0, 1], [0, 1], '--', color='#AFABAB', lw=1)
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('1 - Specificity', fontsize=14)
plt.ylabel('Sensitivity', fontsize=14)
plt.legend(loc="lower right", fontsize=11)
plt.tight_layout()

plt.savefig('img_plot/ROC_curve.png', dpi=300)
plt.savefig('img_plot/ROC_curve.svg', bbox_inches='tight')
plt.savefig('img_plot/ROC_curve.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
